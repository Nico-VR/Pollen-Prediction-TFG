# 1 Introduction

The following notebooks seek to analyze how zero-shot Time Series Foundation Models (zero-shot TSFMs) perform for predicting pollen.

## 1. Dataset
We have a dataset for Alnus glutinosa with the following daily data from 1993 to 2023: pollen, rain, tmax, tmin, tmed.

## 2. Forecasting models
In the following notebooks, what we are going to do is a **benchmark** with which we can evaluate the performance difference between using **models that need training**: ```NHiTS```, ```NBeats``` and using **zero shot models**: ```Moirai```, ```Chronos``` for pollen prediction. For the benchmark, **we analyze how well all the models predict the main pollen season (MPS)** using the two most common methods for calculating it: by percentage and by threshold.

## 3. Estimate Main Pollen Season 
The **main pollen season** (MPS) refers to the core period of the year when the majority of airborne pollen is present.

## 4. Analyzing the MPS with real data
The Alnus glutinosa tree has its own pollen season. The main pollen season in recent years has spanned from late December to mid-January through late February and mid-March in Ourense. Therefore, to be conservative, we can say that **the Alnus pollen season extends from mid-December to mid-March**. **Pollen is absent for most of the year**, more than 250 days a year. Some values are unusually high (e.g., 171, 261, 379, etc.) but **outliers occur seldom in a pollen season**.
